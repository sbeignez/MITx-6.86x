### 13.3.Q1 Counting Dimensions of Feature Vectors

Let $ x \in \mathbb{R}^{150} $, i.e. $ x = [x_1, x_2, ..., x_{150}] $

Dimensions = $$ \sum_{k=1}^p \begin{pmatrix} n + k - 1 \\ k \end{pmatrix} $$

```python
import math
n = 150
p = 3
sum = 0
for k in range(1,p+1):
    sum += math.comb(n + k -1, k)
print("x dimensions:", n)
print("Order:", p)
print("phi(x) dimensions:", sum)
```

### How to plot a function of two variables with matplotlib
https://glowingpython.blogspot.com/2012/01/how-to-plot-two-variable-functions-with.html

```python
import numpy as np
import pylab as pl
```

```python
def z(x1, x2):
    return x1**2 + x1 * x2 + x2**2
```

```python
x1 = np.arange(-3.0,3.0,0.1)
x2 = np.arange(-3.0,3.0,0.1)
X1, X2 = np.meshgrid(x1, x2)
Z = z(X1,X2)
```

```python
im = pl.imshow(Z,cmap=pl.cm.RdBu)
cset = pl.contour(Z,np.arange(-3,3,0.5),linewidths=2,cmap=pl.cm.Set2)
pl.clabel(cset,inline=True,fmt='%1.1f',fontsize=10)
pl.show()
```

## 13.7.Q1Calculating the Radial Basis Kernel

$$ K(x,x') = \exp{(-1/2*||x - x'||^2)} $$

```python
def K(x1, x2):
    return math.exp(-0.5 * math.pow(np.linalg.norm(x1 - x2),2))




print(K(0,0))
```
